#### 1.4.4  Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('IMG_6504.JPG')
rows,cols,ch = img.shape
pts1 = np.float32([[170,190],[760,196],[808,1096],[100,1080]])
pts2 = np.float32([[0,0],[400,0],[400,600],[0,600]])
M =  cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(400,600))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```

```python
img=cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[180,240],[820,220],[820,500],[180,500]])
pts2 = np.float32([[0,0],[820,0],[820,500],[0,600]])
M =  cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(830,510))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```

```python
img=cv2.imread('hero.png')
kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]),plt.yticks([])
plt.show()
```
